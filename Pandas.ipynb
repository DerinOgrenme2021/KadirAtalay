```python
df2=cf_housing_df[cf_housing_df["median_house_value"]>150000]
df2
```

```python
df_100_200 = cf_housing_df[100:200]
df_100_200
```

```python
df_100_200[["longitude","latitude","median_income","median_house_value"]]
```

```python
df_agg =df_100_200.groupby("housing_median_age").agg(["min"])
df_agg
```

```python
df3 = cf_housing_df[cf_housing_df["total_bedrooms"]<600]
df4 = df3[df3["median_house_value"]>150000]
```

```python
df_100_200=df4[0:200]
df_100_200[["housing_median_age","total_rooms","total_bedrooms","population","median_house_value"]]
```

```python
df_100_200.describe()
```

```python
df_agg =df_100_200.groupby("housing_median_age").agg(["max"])
df_agg
```
